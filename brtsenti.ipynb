```python
#bertsenti
from transformers import pipeline

# 1. Create the pipeline
#automatically downloads a model fine-tuned on the SST-2 (Stanford Sentiment Treebank) dataset.
# This defaults to a DistilBERT model, which is a faster, lighter version of BERT
sentiment_pipeline = pipeline("sentiment-analysis")

# 2. Define inputs
data = [
    "I absolutely loved the movie! The plot was brilliant.",
    "The food was cold and the service was terrible."
]

# 3. Predict
results = sentiment_pipeline(data)

# 4. Show results
for text, result in zip(data, results):
    print(f"Text:  {text}")
    print(f"Label: {result['label']}")
    print(f"Score: {result['score']:.4f}")
    print("-" * 30)

import torch
from transformers import BertTokenizer, BertForSequenceClassification
import torch.nn.functional as F

# Load model
model_name = "textattack/bert-base-uncased-SST-2"
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertForSequenceClassification.from_pretrained(model_name)

# Input & Tokenize
text = ["Today was a pleasent day.",
        "This is the worst experience I ever had."]
inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True)

# Predict
with torch.no_grad():
    logits = model(**inputs).logits

# Get result
label_map = {0: "NEGATIVE", 1: "POSITIVE"}
predicted_ids = torch.argmax(logits, dim=1)

for i, pred_id in enumerate(predicted_ids):
    print(f"Text: {text[i]}")
    print(f"Sentiment: {label_map[pred_id.item()]}")
    print("-" * 30)
```
