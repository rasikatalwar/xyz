```python
from transformers import pipeline

qa = pipeline("question-answering", model="deepset/bert-base-cased-squad2")

context = """
    BERT (Bidirectional Encoder Representations from Transformers) was developed
    by Google AI Language team in 2018. It is pre-trained on a large corpus
    of English text using two tasks: Masked Language Modeling and Next Sentence Prediction.
"""

questions = [
    "Who developed BERT?",
    "When was BERT created?",
    "What tasks is BERT pre-trained on?"
]

for q in questions:
    result = qa(question=q, context=context)
    print(f"Q: {q}")
    print(f"A: {result['answer']}  (score: {result['score']:.2f})\n")
```
