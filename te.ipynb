```python
#te
import spacy

# 1. Load the pre-trained English model
# (Run: !python -m spacy download en_core_web_sm in a cell first if missing)
nlp = spacy.load("en_core_web_sm")

# 2. Input Text
text = "The 1st meeting will be held on Monday at 9:00 AM for 2 hours."

# 3. Predict and Filter
doc = nlp(text)
for ent in doc.ents:
    if ent.label_ in ["DATE", "TIME"]:
        print(f"[{ent.label_}] -> {ent.text}")
```
