```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import torch

# 1. Load a standard Encoder-Decoder model (BART)
model_name = "facebook/bart-base"
tokenizer  = AutoTokenizer.from_pretrained(model_name)
model      = AutoModelForSeq2SeqLM.from_pretrained(model_name)

# Input
input_text = "The quick brown fox jumps over the lazy dog."
print("Input text: ")
print(input_text)

# 2. The "Encoder" Phase: Processing input
inputs = tokenizer(input_text, return_tensors="pt")
print("Encodings: ")
print(inputs)

# 3. The "Decoder" Phase: Generating output
# model.generate handles the loop between encoder and decoder automatically
output_ids = model.generate(inputs["input_ids"], max_length=50)

# 4. Convert back to text
print("Decoded output:")
print(tokenizer.decode(output_ids[0], skip_special_tokens=True))
```
