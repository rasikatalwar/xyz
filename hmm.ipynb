```python
import nltk
from nltk.corpus import treebank
from nltk.tag import hmm
import matplotlib.pyplot as plt
import seaborn as sns

nltk.download('treebank')

# Train NLTK HMM tagger
train_data = treebank.tagged_sents()
trainer = hmm.HiddenMarkovModelTrainer()
tagger = trainer.train_supervised(train_data)

sentence = 'Nature is source of life'.split()
tagged = tagger.tag(sentence)
print("Tagged sequence:")
print(tagged)

# Extract words and tags
words = [w for (w, t) in tagged]
tags = [t for (w, t) in tagged]

# Map each tag to an integer "state index"
unique_tags = sorted(set(tags))
tag_to_idx = {tag: i for i, tag in enumerate(unique_tags)}
hidden_states_predicted = [tag_to_idx[t] for t in tags]

print("Most Likely Hidden States (Viterbi):")
print([unique_tags[i] for i in hidden_states_predicted])

# Plot
sns.set_style("darkgrid")
plt.figure(figsize=(8, 4))
plt.plot(hidden_states_predicted, "-o")
plt.yticks(range(len(unique_tags)), unique_tags)
plt.xticks(range(len(words)), words)
plt.title("Predicted POS Tag Sequence (HMM)")
plt.xlabel("Time Step (word position)")
plt.ylabel("Hidden State (POS tag)")
plt.tight_layout()
plt.show()
```
