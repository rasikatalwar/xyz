```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

# 1. Load model
name = "Babelscape/rebel-large"
tokenizer = AutoTokenizer.from_pretrained(name)
model = AutoModelForSeq2SeqLM.from_pretrained(name)

# 2. Input text
text = "Elon Musk lives in Austin, Texas."

# 3. Tokenize and Generate
inputs = tokenizer(text, return_tensors="pt", truncation=True, max_length=512)
output_ids = model.generate(
    **inputs,
    max_length=512,
    num_beams=4,
    early_stopping=True
)

# 4. Decode
extraction = tokenizer.decode(output_ids[0], skip_special_tokens=False)
print(extraction)

# 5. Parse into triplets
triplets = []

parts = extraction.split("<triplet>")

triplets = []

for part in parts:
    if "<subj>" in part and "<obj>" in part:
        s = part.split("<subj>")
        subject  = s[0].strip()
        obj      = s[1].split("<obj>")[0].strip()
        relation = s[1].split("<obj>")[1].strip()
        triplets.append((subject, relation, obj))

# 6. Print results
print("Extracted Relationships:")
print("-" * 40)
for s, r, o in triplets:
    print(f"Subject  : {s}")
    print(f"Relation : {r}")
    print(f"Object   : {o}")
    print("-" * 40)
```
